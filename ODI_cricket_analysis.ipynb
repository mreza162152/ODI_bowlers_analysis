**Changing Columns name**

```python
col_names = ["Players Name", "Playing Span", "Matches Played", "Innings Bowled","Balls Bowled", "Runs Conceded", 
             "Wickets Taken", "Bowling Average", "Economy Rate", "Bowling Strike Rate", "Four Wickets in a Innings",
             "Five Wickets in a Innings"]

# df_bowler = pd.read_excel("../ODI_cricket.xlsx", sheet_name = "bowler", engine= "openpyxl")
df_bowler.columns = col_names
display(df_bowler)
```

```python
df_bowler[["Players Name", "Country"]] = df_bowler["Players Name"].str.split("(", expand=True)
display(df_bowler)
```

```python
df_bowler['Country'] = df_bowler['Country'].str.replace(")", "")
display(df_bowler.head())
```

```python
df_bowler["Country"].value_counts()
```

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
def asia_check(x):
    if "Asia" in x:
        return "Yes"
    else:
        return "No"
```

```python
def africa_check(x):
    if "Afr" in x:
        return "Yes"
    else:
        return "No"
```

```python
df_bowler["Played for ICC"] = df_bowler["Country"].apply(icc_check)
df_bowler["Played for ASIA"] = df_bowler["Country"].apply(asia_check)
df_bowler["Played for AFRICA"] = df_bowler["Country"].apply(africa_check)

display(df_bowler.head(10))
```

```python
print("Total player played in ICC are :",list(df_bowler["Played for ICC"]).count("Yes"))
```

```python
df_bowler["Country"] = df_bowler["Country"].str.replace("ICC", "")
df_bowler["Country"] = df_bowler["Country"].str.replace("Asia", "")
df_bowler["Country"] = df_bowler["Country"].str.replace("Afr", "")
df_bowler["Country"] = df_bowler["Country"].str.replace("/", "")

display(df_bowler.head(10))
```

```python
df_bowler["Country"].value_counts()
```

```python
print(len(list(df_bowler['Country'].value_counts())), " Countries are playing")
```

**Player who had longest and shortest period of time**

```python
df_bowler[["Starting Year", "Ending Year"]] = df_bowler["Playing Span"].str.split("-", expand=True)
df_bowler = df_bowler.drop("Playing Span", axis = 1)

display(df_bowler.head())
```

```python
df_bowler.info()
```

```python
df_bowler["Starting Year"] = df_bowler["Starting Year"].astype('int') 
df_bowler["Ending Year"] = df_bowler["Ending Year"].astype('int')

print(df_bowler.info())
```

```python
df_bowler["Year Played"] = df_bowler["Ending Year"] - df_bowler["Starting Year"]
df_bowler = df_bowler.drop(["Starting Year", "Ending Year"] , axis = 1)
display(df_bowler.head())
```

```python
display(df_bowler[df_bowler["Year Played"] == df_bowler["Year Played"].max()])
```

```python
display(df_bowler[df_bowler["Year Played"] == df_bowler["Year Played"].min()])
```
